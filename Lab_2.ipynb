# Question 1.1 (Exercise 5 CH2)

```python
def sense(x):
    return x
```

```python
def simulate(Δt, x, dx):
    x += Δt * dx
    return x
```

```python
def control(t, y, x_d):
    k = 1
    ux = k * (x_d[0] - y[0])
    uy = k * (x_d[1] - y[1])
    return array([ux, uy])
```

```python
tf = 5.0
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)


x = np.array([1.0, 2.0])
x_d = np.array([3., 3.])
x_log = [np.copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y, x_d)    
    x = simulate(Δt, x, u)
    x_log.append(np.copy(x))
    
x_log = np.array(x_log)
```

```python
grid()
plt.axis('equal')
plot(x_log[:,0], x_log[:,1])
```

# Question 2.4 (Exercise 3 CH 3)

```python
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    m = 0.8
    g = 9.8
    dx = np.array([x[3],x[4],x[5], u[0]/m,u[1]/m, u[2]])
    dx[5] = (dx[5] - m * g) / m
    x += Δt * dx
    return x
```

```python
def control(t, y, x_d, e):
    k_p = 2
    k_d = 2
    k_i = 0.01
    m = 0.8
    g = 9.8
    ux = 0
    uy = 0
    e += (x_d[2] - y[2])
    uz = k_p * (x_d[2] - y[2]) + k_d * (x_d[5] - y[5]) + m * g + k_i * e * 0.1
    return [array([ux, uy, uz]), e]
```

```python
tf = 5.0
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)


x = np.array([0,0,10.0, 0, 0, 0])
x_d = np.array([0,0,15.0, 0,0,0])

x_log = [np.copy(x)]
e = 0.0

for t in time:
    y = sense(x)
    res = control(t, y, x_d, e)
    u = res[0]
    e = res[1]
    x = simulate(Δt, x, u)
    x_log.append(np.copy(x))

x_log = np.array(x_log)
```

```python
grid()
plot(time, x_log[:-1,2])
```
