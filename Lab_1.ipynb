# Exercise 2

```python
def sense(x):
    return x
```

```python
def simulate(Δt, x, dx):
    x += Δt * dx
    return x
```

```python
def control(t, a):
    ux = a * cos(t)
    uy = -a * (sin(t)**2 - cos(t)**2)
    return array([ux, uy])
```

```python
a = 2
```

```python
tf = 2 * np.pi / 1
Δt = 0.1    # Time step
time = np.linspace(0.,tf, int(tf / Δt) + 1)


x = np.array([-0.1, -0.1])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, a)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)

grid()
plot(x_log[:,0], x_log[:,1])
```

# Exercise 4

```python
def sense(x):
    return x
```

```python
def simulate(Δt, x, dx, wind):
    x += Δt * (dx + wind)
    return x
```

```python
def control(t):
    ux = -sin(t)
    uy = cos(t)
    return array([ux, uy])
```

```python
wind = np.array([0.1,0.1])
```

```python
np.random.seed(0)
wind_mean = [0.0, 0.0]
wind_std = [0.1, 0.1]
wind_random = np.random.normal(wind_mean, wind_std)
```

```python
# Standard Wind
tf = 20
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([2., 1.])
x_random = array([2.0, 1.0])
x_log_standard = [copy(x)]
x_log_random = [copy(x)]

for t in time:
    y = sense(x)
    y_random = sense(x_random)
    u = control(t)    
    x = simulate(Δt, x, u,wind)
    x_random = simulate(Δt, x_random, u,wind_random)
    x_log_standard.append(copy(x))
    x_log_random.append(copy(x_random))
x_log_standard = array(x_log_standard)
x_log_random = array(x_log_random)

grid()
plot(x_log_standard[:,0], x_log_standard[:,1],label = 'Stable Wind')
plot(x_log_random[:,0], x_log_random[:,1], label = 'Random Wind')
plt.legend()
```
