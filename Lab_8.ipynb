```python
import matplotlib.pyplot as plt
import numpy as np
u = [157, 179, 202, 208, 125, 202, 171, 193, 187, 175]
angle = [-0.75, 4.75, 10.5, 12.0, -8.75, 10.5, 2.75, 8.25, 6.75, 3.75]
sb = [9676.89, 12076.3, 17203.4, 6939.78, 14102.6, 8824.73, 4071.5, 14102.6, 15393.8, 12667.7]
distance = [1.25755, 1.12571, 0.943163, 1.48498, 1.0417, 1.31687, 1.93872, 1.0417, 0.997058, 1.09912]
```

```python
sb = [1/x for x in sb]
```

```python
coefficients = np.polyfit(sb, distance, 1)
fit_line = np.poly1d(coefficients)
```

```python
plt.scatter(sb, distance)
plt.xlabel('1/s')
plt.ylabel('d')
plt.plot(sb, fit_line(sb), color='red', label='Fit Line')
```

```python
coefficients = np.polyfit(u, angle, 1)
fit_line = np.poly1d(coefficients)
```

```python
plt.scatter(u, angle)
plt.xlabel('u')
plt.ylabel('angle')
plt.plot(u, fit_line(u), color='red', label='Fit Line')
```

```python
def approximation(sb, distance, u, angle, s_in, u_in):
    coefficients_s = np.polyfit(sb, distance, 1)
    fit_line_s = np.poly1d(coefficients_s)
    predict_d = fit_line_s(1/s_in)
    
    coefficients_u = np.polyfit(u, angle, 1)
    fit_line_u = np.poly1d(coefficients_u)
    predict_angle = fit_line_u(u_in)
    
    return predict_d, predict_angle
```

```python
predict_d,predict_angle = approximation(sb, distance, u, angle,12076.3, 170)
print("Predict Distance: ", predict_d)
print("Predict Angle: ", predict_angle)
```
